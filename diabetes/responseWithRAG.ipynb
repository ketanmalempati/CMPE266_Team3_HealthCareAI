```
from langchain.embeddings import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
```

```
from langchain.vectorstores import Chroma
!pip install chromadb
```

```
vectordb = Chroma(persist_directory="/content/drive/MyDrive/266/asthma/data", embedding_function=embeddings)
```

```
def getcontext(query):
  docs = vectordb.similarity_search(query,k=5)
  return docs
```

```
from together import Together

def greet(question,sliderdata):
  print(sliderdata)
  client = Together(api_key="f96d39397eef84c3fc286f3ce491f429dbd77232adfb79fc893b5a0bce9c819b")
  context ="Gregangelo museum (Into the Rabbit Hole tour) Musée Mécanique Audium SF crosstown trail I particularly like urban hikes (like the crosstown trail). It gives you an opportunity to see multiple neighborhoods and explore different parts of the city. Make your own by hopping on a muni train, get off near the end of the line, and start exploring!"
  context =getcontext(question)[0].page_content
  response = client.chat.completions.create(
      model="meta-llama/Llama-3-8b-chat-hf",
      messages=[{"role": "user","content": "[INST] Answer the question in detail based on the following context: "+context+ " Question: "+question+"[/INST]"}],
  )
  return(response.choices[0].message.content)
```

```
import gradio

gradio.Interface(greet, "text", "text").launch(share=True,debug=True)
```

```
import os
from together import Together

client = Together(api_key="f96d39397eef84c3fc286f3ce491f429dbd77232adfb79fc893b5a0bce9c819b")
question = 'what are the trigger factors in asthma?'
response = client.chat.completions.create(
    model="meta-llama/Llama-3-8b-chat-hf",
    messages=[{"role": "user", "content": question}],
)
print("meta-llama/Llama-3-8b-chat-hf: ", response.choices[0].message.content)
response = client.chat.completions.create(
    model="google/gemma-7b-it",
    messages=[{"role": "user", "content": question}],
)
print("google/gemma-7b-it: ", response.choices[0].message.content)
```
